```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
DATA_SOURCE='https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
df_t=pd.read_html(DATA_SOURCE)
df_m=df_t[0]
```

```python
UNIVERSE=df_m['Symbol'].values
```

```python
UNIVERSE=list(UNIVERSE)
```

```python
DATA=yf.download(UNIVERSE,interval="1d",start="2013-06-06",end="2022-06-06")
```

```python
len(DATA)
```

```python
close=DATA['Adj Close']
len(close)
```

```python
close.head()
len(close)
```

```python
close=close.dropna(axis=1)
len(close)
```

```python
len(close)
```

```python
returns=close.pct_change()
```

```python
returns
```

```python
close
```

```python
ts_delta_3=close.diff(periods=3,axis=0)
```

```python
ts_delta_3
```

```python
rank=ts_delta_3.sub(ts_delta_3.min(axis=1),axis=0)
rank=rank.div(rank.max(axis=1),axis=0)
```

```python
rank
```

```python
rank=rank.sub(rank.mean(axis=1),axis=0)
```

```python
rank
```

```python
alpha=rank*returns
```

```python
alpha
```

```python
HEDGE_FUND_CASH_TRADED=20000000
alpha=alpha*HEDGE_FUND_CASH_TRADED
alpha
```

```python
PnL=alpha.sum(axis=1)
```

```python
PnL
```

```python
plthelp=[]
for i in range(len(PnL)):
    plthelp.append(i)

plthelp=np.array(plthelp)
```

```python
plt.plot(plthelp,PnL)
```

```python
#The graph shows the PnL generated per Day over a span of 8 years
#On Profit making days the excess in 20 Million is collected as profit
#On loss making days the deficit is made up by reinjecting cash into the portfolio
```
